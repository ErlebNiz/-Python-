### Задание 3

```python
df_path3 = os.path.normpath(r'C:\Users\XE\Downloads\ml-latest-small\ratings.csv')
```

```python
df_ratings = pd.read_csv(df_path3)
```

```python
df_ratings.head(10)
```

```python
df_ratings['timestamp'] = pd.to_datetime(df_ratings['timestamp'], unit='s')
```

```python
df_ratings.head(10)
```

```python
user_lifetime = df_ratings.groupby('userId').agg(
    min_time=('timestamp', 'min'),
    max_time=('timestamp', 'max'),
    num_ratings=('rating', 'count')
)
```

```python
user_lifetime.head(10)
```

```python
user_lifetime['lifetime'] = (user_lifetime['max_time'] - user_lifetime['min_time']).dt.days
```

```python
active_users = user_lifetime[user_lifetime['num_ratings'] > 100]
```

```python
active_users
```

```python
average_lifetime = active_users['lifetime'].mean()
```

```python
round(average_lifetime, 2)
```

```python
super_active_users = user_lifetime[(user_lifetime['lifetime'] == 0) & 
                                   (user_lifetime['num_ratings'] > 100)].reset_index()['userId'].count()
```

```python
super_active_users
```
