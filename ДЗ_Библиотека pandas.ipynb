### Задание 3

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://pythonworld.ru/tipy-dannyx-v-python/stroki-funkcii-i-metody-strok.html'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

# Находим таблицу с классом 'docutils'
table = soup.find('table', class_='docutils')

if table:
    # Создаем список для хранения данных
    data = []
    
    # Извлекаем строки таблицы
    rows = table.find_all('tr')
    
    for row in rows:
        cells = row.find_all(['td', 'th'])  # Извлекаем как ячейки данных, так и заголовки
        row_data = [cell.get_text(strip=True) for cell in cells]  # Получаем текст и убираем лишние пробелы
        data.append(row_data)  
else:
    print("Таблица не найдена.")  
```

```python
df = pd.DataFrame(data)
```
